```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
```

```python
website = 'https://www.zoopla.co.uk/to-rent/flats/london/'

url = requests.get(website)
```

```python
def page_scroller(url:str) -> None:
    total_pages = 5
    current_page = 1
    
    while current_page <= total_pages:
        print(f'{url}?price_frequency=per_month&price_max=2000&q=London&results_sort=newest_listings&search_source=to-rent&pn={current_page}')
        current_page += 1
        
```

```python
soup = bs(url.text, 'html.parser')
```

```python
london_data = pd.DataFrame(columns=['Address','Distance','Rent','Bedrooms','Date Listed'])


for row in soup.find('div', class_='css-8jz4jb-SearchResultsLayoutGroup ed9e91o8').find_all('div', class_='css-mww4lt-StyledContent e9kuaf127'):
    address = row.find('p', class_='css-1uvt63a-Address e9kuaf13').text
    distance = row.find('p', class_='css-nwapgq-Text eczcs4p0').text
    rent = row.find('p', class_='css-xz7r6w-Price e9kuaf15').text
    bedrooms = row.find('p', class_='css-r8a2xt-Text eczcs4p0').text
    datelisted = row.find('span', class_='css-55epu1-FooterText e9kuaf12')
    
    london_data = london_data.append({'Address':address, 'Distance':distance, 'Rent':rent, 'Bedrooms':bedrooms, 'Date Listed':datelisted}, ignore_index=True)

london_data.head(200)
```

```python
page_scroller(soup.text)
```
